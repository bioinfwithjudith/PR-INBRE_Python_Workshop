###Exercise 3.3.3: Create a Panda DataFrame from a Nested Dictionary

```python
#Use this block of code for notes
crtam_dictionary = {'CRTAM_HUMAN':{'length':393,'organism':'human'},
                    'CRTAM_MOUSE':{'length':781,'organism':'mouse'},
                    'CADMI_HUMAN':{'length':442,'organism':'human'}}
crtam_dictionary
```

```python
df_from_dictionary = pd.DataFrame(data=crtam_dictionary)
df_from_dictionary
```

###Exercise 3.3.4: Let's fix our new Pandas DataFrame

```python
# let's transpose this table to have my samples as the rows
df_transpose = df_from_dictionary.T
df_transpose
```

```python
# my index are my gene names, but I want them to be a aprt of the data under a column named gene
df_transpose_fixed_index = df_transpose.reset_index()
df_transpose_fixed_index
```

```python
final_df = df_transpose_fixed_index.rename(columns={'index':'gene'})
final_df
```

###Exercise 3.3.5: Export a Panda DataFrame as a CSV File

```python
final_df.to_csv('crtam_data.csv',sep=',')
```

#Proficiency Assessment

```python
#open our file of interest
fasta_file = open('crtam.faa','r')

#start empty lists (these list we will fill up in a for loop then added to a dataframe)
gene_name_list=[]
protein_length_list=[]
dna_length_list=[]
RSNNEETS_list=[]
PGLRLRLLLL_list=[]
KAH_list=[]

#go through each line and try to understand what each line is doing
for fasta_line in fasta_file:
  if fasta_line[0] == '>':
    gene_name = fasta_line[1:].strip('\n')
    gene_name_list.append(gene_name)
  elif fasta_line[0] != '>':
    protein_length = len(fasta_line)
    protein_length_list.append(protein_length)
    dna_length = len(fasta_line)*3
    dna_length_list.append(dna_length)
    if 'RSNNEETS' in fasta_line:
      RSNNEETS = "True"
      print(RSNNEETS)
      RSNNEETS_list.append(RSNNEETS)
    elif 'PGLRLRLLLL' in fasta_line:
      PGLRLRLLLL = "True"
      PGLRLRLLLL_list.append(PGLRLRLLLL)
    elif 'KAH' in fasta_line:
      KAH = "True"
      KAH_list.append(KAH)
fasta_file.close()
```

```python
#make a nested list of our newly created lists
list_of_lists = [gene_name_list,protein_length_list,dna_length_list,RSNNEETS_list,PGLRLRLLLL_list,KAH_list]
list_of_lists
```

```python
#create a dataframe from the list of lists
df = pd.DataFrame(list_of_lists)
df
```

```python
#let's transpose our dataframe because we have three samples and we want the samples to be rows
df_transpose = df.T
df_transpose
```

```python
#we can also rename our columns
df_change_column_names = df_transpose.rename(columns={0:'gene',1:'protein_length',2:'dna_length',3:'RSNNEETS',4:'PGLRLRLLLL',5:'KAH'})
df_change_column_names
```

```python
#export a csv file from our final dataframe
df_change_column_names.to_csv('data.csv',sep=',')
```
